# 4. Build a collaborative filtering model using SVD. You can use SVD from surprise or build it from scratch(Note: Incase you’re building it from scratch you can limit your data points to 5000 samples if you face memory issues). Build a collaborative filtering model using kNNWithMeans from surprise. You can try both user-based and item-based model

```python
from surprise import KNNWithMeans
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import Dataset,Reader
from surprise import SVD
from surprise import accuracy
```

```python
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(ratings[['author', 'product','score']],reader)
```

```python
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=0.25,random_state=456)
```

```python
testset
```

```python
svd_model = SVD(n_factors=5,biased=False)
svd_model.fit(trainset)
```

```python
test_pred = svd_model.test(testset)
```

```python
accuracy.rmse(test_pred)
```

KNN item based model

```python
algo = KNNWithMeans(k=5, sim_options={'user_based': False})
```

```python
algo.fit(trainset)
```

```python
test_pred1 = algo.test(testset)
```

```python
print("Item-based Model : Test Set")
accuracy.rmse(test_pred1, verbose=True)
```

User based model

```python
algo_i = KNNWithMeans(k=5, sim_options={'user_based': True})
```

```python
algo_i.fit(trainset)
```

```python
test_pred2 = algo_i.test(testset)
```

```python
print("Item-based Model : Test Set")
accuracy.rmse(test_pred2, verbose=True)
```

# 11. In what business scenario you should use CF based Recommendation Systems ?

CF recommendation is recommendation based on the other people with same interest.

For example Person 1 wound have watched movie A,B,C
Person 2 would have watched movie B,C
